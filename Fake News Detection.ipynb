```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report 
import re
import string
```

```python
data_fake = pd.read_csv('Fake.csv')
data_true = pd.read_csv('True.csv')

data_fake.head()
```

```python
data_true.head()
```

```python
data_fake['class'] = 0
data_true['class'] = 1


data_fake.shape , data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480,23470,-1):
    data_fake.drop([i] , axis = 0 , inplace = True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416,21406,-1):
    data_true.drop([i] , axis = 0 , inplace = True)
```

```python
data_fake.shape , data_true.shape
```

```python
data_fake_manual_testing['class'] = 0
data_true_manual_testing['class'] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake,data_true],axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title','subject','date'],axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'],axis = 1, inplace = True)
data.columns
```

```python
data.head(10)
```

```python
def wordopt(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\[.*?\]', '', text)  # Remove content inside square brackets
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub(r'\n', ' ', text)  # Replace newline with space
    text = re.sub(r'\w*\d\w*', '', text)  # Remove words containing digits
    text = re.sub(r'\s+', ' ', text).strip()  # Normalize whitespace
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)  # Learn vocabulary and transform training data
xv_test = vectorization.transform(x_test)        # Only transform test data
```

```python
from sklearn.linear_model import LogisticRegression 

lr = LogisticRegression()
lr.fit(xv_train , y_train)
```

```python
pred_lr = lr.predict(xv_test)

lr.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xv_train , y_train)
```

```python
pred_dt = dt.predict(xv_test)
```

```python
dt.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state = 0)
gb.fit(xv_train , y_train)
```

```python
pred_gb = gb.predict(xv_test)
```

```python
gb.score(xv_test , y_test)
```

```python
print(classification_report(y_test , pred_gb))
```
